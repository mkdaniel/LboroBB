# Extract Data

read a year long Loughborough pyranometer data text file and extract to individual day files for easier processing
* the year long file is large and will easily eat all available memory

NB, data comes from a PostgreSQL database, this process can be adapted to read data directly from that DB easily enough.

Data is expected to be csv file with columns
* MTR TimeStamp,
* GHI [W/m2],
* DHI [W/m2],
* BNI [W/m2],
* G_S34 [W/m2],
* G_S45 [W/m2],
* Tamb [deg.C],
* RH [%],
* Wind Dir [deg.bearing],
* Wind Speed [m/s]
